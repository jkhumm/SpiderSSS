```python
df_problems['lst_topics'] = df_problems['topicSlug'].apply(lambda x:x[1:-1].replace('\'','').replace(' ','').split(','))
df_problems['topicNum'] = df_problems['lst_topics'].apply(lambda x:len(x))
print('一个题目标签数量最多为：',df_problems['topicNum'].max())
```

```python
sns.distplot(df_problems['topicNum'])
plt.title('每个题目拥有标签数量分布')
```

```python
# 5个标签
for num in range(5):
    df_problems['topic_'+str(num+1)] = df_problems['lst_topics'].apply(lambda x:x[num] if len(x)>num and x[num] != '' else None)
    df_problems['topic_'+str(num+1)] = df_problems['topic_'+str(num+1)].map(dict_topic)
```

### 标签数量统计

```python
plt.figure(figsize=(20,5))
sns.countplot([dict_topic[topic] for topic in df_problems['lst_topics'].sum() if topic!=''])
plt.title('标签总量统计')
plt.xticks(rotation=90)

for i in range(5):
    plt.figure(figsize=(20,5))
    sns.countplot(df_problems['topic_' + str(i+1)])
    plt.title('标签'+str(i+1)+'数量统计')
    plt.xticks(rotation=90)
```

### 标签组合数量统计

```python
# 组合标签
seq = ' - '
df_problems['topic_1_2'] = df_problems['topic_1'] + seq + df_problems['topic_2']
df_problems['topic_1_3'] = df_problems['topic_1'] + seq + df_problems['topic_3']
df_problems['topic_2_3'] = df_problems['topic_2'] + seq + df_problems['topic_3']
df_problems['topic_1_2_3'] = df_problems['topic_1'] + seq + df_problems['topic_2'] + seq + df_problems['topic_3']

plt.figure(figsize=(10,5))
gp_topic_1_2 = df_problems.groupby(['topic_1_2'])['questionId'].count().reset_index()
gp_topic_1_2 = gp_topic_1_2.sort_values(by=['questionId'], ascending=False)[:20]
sns.barplot(y='topic_1_2',x='questionId',data=gp_topic_1_2)
plt.title('标签1-标签2 组合数量统计(top20)')

plt.figure(figsize=(10,5))
gp_topic_1_3 = df_problems.groupby(['topic_1_3'])['questionId'].count().reset_index()
gp_topic_1_3 = gp_topic_1_3.sort_values(by=['questionId'], ascending=False)[:20]
sns.barplot(y='topic_1_3',x='questionId',data=gp_topic_1_3)
plt.title('标签1-标签3 组合数量统计(top20)')

plt.figure(figsize=(10,5))
gp_topic_2_3 = df_problems.groupby(['topic_2_3'])['questionId'].count().reset_index()
gp_topic_2_3 = gp_topic_2_3.sort_values(by=['questionId'], ascending=False)[:20]
sns.barplot(y='topic_2_3',x='questionId',data=gp_topic_2_3)
plt.title('标签2-标签3 组合数量统计(top20)')

plt.figure(figsize=(10,5))
gp_topic_1_2_3 = df_problems.groupby(['topic_1_2_3'])['questionId'].count().reset_index()
gp_topic_1_2_3 = gp_topic_1_2_3.sort_values(by=['questionId'], ascending=False)[:20]
sns.barplot(y='topic_1_2_3',x='questionId',data=gp_topic_1_2_3)
plt.title('标签1-标签2-标签3 组合数量统计(top20)')
```
